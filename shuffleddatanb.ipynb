```python
# working with dataframes
import pandas as pd
import numpy as np

# working with paths for files
import pathlib, glob
from pathlib import Path

import os
```

```python
path = Path("../1015finalproject/shuffleddata.csv")
df = pd.read_csv(path)
```

```python
df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X_encoded = pd.get_dummies(df['micro_prev_organism'])
y_encoded = pd.get_dummies(df['resist'])

is_train = df['is_train'] == 1
X_train = X_encoded[is_train]
y_train = y_encoded[is_train]

X_test = X_encoded[~is_train]
y_test = y_encoded[~is_train]

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("y_pred length", len(y_pred))

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
